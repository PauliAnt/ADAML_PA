```python
import pandas as pd
import numpy as np

df = pd.read_csv('datasets_56102_107707_monthly-beer-production-in-austr.csv',index_col=0,parse_dates=True)
values = df.values.flatten()


# Partition 70 % train data, 15% val data, 15% test data
# Change this to alter data size
seq_length = 12

# train data
train_partition = 0.7
train_breakpoint = int(train_partition*len(values))
inds = range(seq_length,train_breakpoint-seq_length)


data = np.zeros((len(inds),seq_length+1))
for ii in inds:    
    data[ii-seq_length,] = values[ii-seq_length:ii+1]


columns = [f'x{ii}' for ii in range(seq_length,0,-1)] + ['y']
df_train = pd.DataFrame(data,index=df.index[inds],columns=columns)
df_train.to_csv('train_data.csv')


# Remove train data
df = df.iloc[train_breakpoint:]
values = values[train_breakpoint:]

# validation data
val_partition = 0.5 # 50% of the remaining data
val_breakpoint = int(val_partition*len(values))
inds = range(seq_length,val_breakpoint-seq_length)


data = np.zeros((len(inds),seq_length+1))
for ii in inds:    
    data[ii-seq_length,] = values[ii-seq_length:ii+1]


columns = [f'x{ii}' for ii in range(seq_length,0,-1)] + ['y']
df_val = pd.DataFrame(data,index=df.index[inds],columns=columns)
df_val.to_csv('val_data.csv')


# Remove validation data
df = df.iloc[val_breakpoint:]
values = values[val_breakpoint:]

# Test data (use rest of the data)
inds = range(seq_length,len(values)-seq_length)

data = np.zeros((len(inds),seq_length+1))
for ii in inds:    
    data[ii-seq_length,] = values[ii-seq_length:ii+1]


columns = [f'x{ii}' for ii in range(seq_length,0,-1)] + ['y']
df_test = pd.DataFrame(data,index=df.index[inds],columns=columns)
df_test.to_csv('test_data.csv')

```
