```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams.update({
    "font.family": "serif ",
    "font.size":23
})

data_df = pd.read_csv('datasets_56102_107707_monthly-beer-production-in-austr.csv', parse_dates=['Month'], index_col='Month' )
data_df
```

```python
data_df.plot( figsize = (15, 4), title='Monthly Beer Production in Australia 1956-1995' )
data_df.iloc[0:(5*12), :].plot( figsize = (15, 4), title='Monthly Beer Production in Australia 1956-1960' )
```

```python
from statsmodels.tsa.seasonal import STL

stl = STL(data_df)
stl_fitted = stl.fit()
stl_fig = stl_fitted.plot()

stl_fig.set_size_inches(15, 11)
for ax in stl_fig.axes:
    ax.grid(True)
    if ax.get_title() == 'Observed':
        ax.set_title('')

stl_fig.tight_layout()

stl_fig.savefig('stl_decomposition.pdf', dpi=300)

```
