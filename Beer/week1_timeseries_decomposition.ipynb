```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams.update({
    "font.family": "serif ",
    "font.size":23
})

data_df = pd.read_csv('datasets_56102_107707_monthly-beer-production-in-austr.csv', parse_dates=['Month'], index_col='Month' )
data_df
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(20, 13))

data_df.plot(ax=axes[0])
axes[0].set_title('Full Data')

data_df.iloc[0:(5*12), :].plot(ax=axes[1])
axes[1].set_title('First 5 Years')

plt.tight_layout()
plt.show()

fig.savefig('data_plot.pdf', dpi=600)

```

```python
from statsmodels.tsa.seasonal import STL

stl = STL(data_df)
stl_fitted = stl.fit()
stl_fig = stl_fitted.plot()

stl_fig.set_size_inches(15, 11)
for ax in stl_fig.axes:
    ax.grid(True)
    if ax.get_title() == 'Observed':
        ax.set_title('')

stl_fig.tight_layout()

stl_fig.savefig('stl_decomposition.pdf', dpi=300)

```
