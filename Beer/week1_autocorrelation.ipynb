```python
import pandas as pd
from statsmodels.graphics.tsaplots import plot_acf
df = pd.read_csv('datasets_56102_107707_monthly-beer-production-in-austr.csv',index_col=0,parse_dates=True)


df.plot()
```

```python
import matplotlib.pyplot as plt
plt.rcParams.update({
    "font.family": "serif ",
    "font.size":22
})



# Example data (replace this with your own time series data)
# np.random.seed(42)
# data = np.random.randn(100)  # Random time series data

# Function to calculate autocorrelation for different lags
def autocorr(x, lag):
    return np.corrcoef(x[:-lag], x[lag:])[0, 1]

# Calculate autocorrelation for desired lags
lags = 50
autocorr_values = [autocorr(df['Monthly beer production'], lag) for lag in range(1, lags+1)]

# Plot autocorrelation function without confidence intervals
plt.figure(figsize=(12, 7))
plt.stem(range(1, lags+1), autocorr_values)
plt.title("Autocorrelation Function")
plt.xlabel("Lag")
plt.ylabel("Autocorrelation")
plt.grid(True)
plt.show()
```
